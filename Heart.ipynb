```python
cat_cols = ['sex','exng','caa','cp','fbs','restecg','slp','thall']
cont_cols =[ "age","trtbps","chol","thalachh","oldpeak", "O2"]

# encoding the categorical columns
df1 = pd.get_dummies(data, columns = cat_cols, drop_first = True)

x = df1.drop(['output'], axis = 1)
y = data['output']
```

# Standardization

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()
x[cont_cols] = sc.fit_transform(x[cont_cols])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.15,random_state=101, shuffle=True)

print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# LOGISTIC REGRESSION
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression  

print("LOGISTIC REGRESSION")
mod = LogisticRegression(random_state=0)     
mod.fit(xtrain, ytrain)  
# mod.fit(xtest, ytest)
y_pred= mod.predict(xtest)  
# print(y_pred)
# print(ytest)
print(mod.score(xtest, ytest))

from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(ytest, y_pred)  
# 0 -> false     1 -> true
# TP   FN
# FP   TN
tp = cm[0][0]
fp = cm[1][0]
fn = cm[0][1]
tn = cm[1][1]
print("Confusion matrix: ",cm)
print("True positive :",cm[0][0])
print("True Negative :",cm[1][1])

print("Accuracy: ",((tp + tn)/(tp + fp + fn + tn))*100)
# calculate prediction Precision = TruePositives / (TruePositives + FalsePositives)
# precision = precision_score(ytest, y_pred)
# print('Precision: %.3f' % precision)
Precision = ((tp)/(tp + fp))
print("Precision : ", Precision)

# calculate recall Recall = TruePositives / (TruePositives + FalseNegatives)
# recall = recall_score(ytest, y_pred, average='binary')
# print('Recall: %.3f' % recall)
Recall = ((tp)/(tp + fn))
print('Recall: ', Recall)

# calculate score F-Measure = (2 * Precision * Recall) / (Precision + Recall)
# score = f1_score(ytest, y_pred, average='binary')
# print('F-Measure: %.3f' % score)

Fscore = (2 * Precision * Recall) / (Precision + Recall)
print("F-score : ", Fscore)


print("MAE : ", mean_absolute_error(ytest,y_pred))
print("MSE : ", mean_squared_error(ytest,y_pred))
```
